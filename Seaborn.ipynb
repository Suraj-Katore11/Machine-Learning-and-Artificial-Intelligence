```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
order_list = pd.read_excel(r"C:\Users\Soft\Desktop\Py\Docs\P6-AmazingMartEU2Geo.xlsx",
                          sheet_name="ListOfOrders")
order_list.head(2)
```

```python
order_break = pd.read_excel(r"C:\Users\Soft\Desktop\Py\Docs\P6-AmazingMartEU2Geo.xlsx",
                           sheet_name="OrderBreakdown")
order_break.head(3)
```

```python
df = order_list.merge(order_break, on="Order ID", how="inner")
df.head()
```

```python
df.info()
```

```python
df.head(2)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(13, 5))
sns.countplot(data=df, x="Country", hue="Category", orient="h")
plt.title("Count of Country")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(13, 5))
# sns.color_palette("mako", as_cmap=True)
sns.histplot(data=df, x="Sales", hue="Category",palette="mako",
             kde=True,log_scale=True)
```

```python
plt.figure(figsize=(13, 5))
sns.histplot(data=df, x="Quantity", hue="Region",palette="mako",
             kde=True)
```

```python
plt.figure(figsize=(13, 5))
# sns.color_palette("mako", as_cmap=True)
sns.histplot(data=df, x="Sales", hue="Region",palette="mako",
             kde=True,log_scale=True)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="Sales", y="Profit", size="Quantity", style="Region",
               legend="full")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Total Sales by Category and Region")
plt.text(-0.5, 500000, "This is demo")
sns.barplot(data=df, x="Category", y="Sales", estimator="sum", hue="Region", palette="mako")
plt.show()
```

```python
sns.boxplot(data=df, y="Sales", log_scale=True, hue="Segment",gap=0.5, fill=False)
plt.legend()
plt.show()
```

```python
sns.jointplot(data=df, x="Sales", y="Profit", hue="Category")
```
