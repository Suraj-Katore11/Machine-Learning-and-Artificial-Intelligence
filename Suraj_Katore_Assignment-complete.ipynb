```python
sku_data['Inventory Turnover'] = sku_data['Units sold'] / (sku_data['Units sold'] + sku_data['Units returned'])
```

```python
inventory_turnover_summary = sku_data.groupby('ASIN').agg({
    'Units sold': 'sum',
    'Units returned': 'sum',
    'Inventory Turnover': 'mean'
}).reset_index().sort_values(by='Inventory Turnover', ascending=False)
```

```python
top_turnover_skus = inventory_turnover_summary.head(5)
bottom_turnover_skus = inventory_turnover_summary.tail(5)
```

```python
top_turnover_skus
```

```python
bottom_turnover_skus
```

```python
# Inventory Turnover Visualization
plt.figure(figsize=(10, 6))
plt.bar(top_turnover_skus['ASIN'], top_turnover_skus['Inventory Turnover'], color='green')
plt.xlabel('ASIN')
plt.ylabel('Inventory Turnover')
plt.title('Top 5 SKUs by Inventory Turnover')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_summary = sku_data.groupby('ASIN').agg({
    'Units sold': 'sum',
    'Sales': 'sum'
}).reset_index()
```

```python
top_skus = sales_summary.sort_values(by='Units sold', ascending=False).head(5)
bottom_skus = sales_summary.sort_values(by='Units sold').head(5)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(top_skus['ASIN'], top_skus['Units sold'], color='green')
plt.title('Top 5 SKUs by Units Sold')
plt.xlabel('ASIN')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
```

```python
plt.subplot(1, 2, 2)
plt.bar(bottom_skus['ASIN'], bottom_skus['Units sold'], color='green')
plt.title('Bottom 5 SKUs by Units Sold')
plt.xlabel('ASIN')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sku_data['Return Rate'] = (sku_data['Units returned'] / sku_data['Units sold']) * 100
return_rate_summary = sku_data.groupby('ASIN').agg({
    'Units sold': 'sum',
    'Units returned': 'sum',
    'Return Rate': 'mean'
}).reset_index().sort_values(by='Return Rate', ascending=False)
top_return_skus = return_rate_summary.head(5)
bottom_return_skus = return_rate_summary.tail(5)
```

```python
top_return_skus
```

```python
bottom_return_skus
```

```python
# Return Rate Visualization
plt.figure(figsize=(10, 6))
plt.bar(top_return_skus['ASIN'], top_return_skus['Return Rate'], color='orange')
plt.xlabel('ASIN')
plt.ylabel('Return Rate (%)')
plt.title('Top 5 SKUs by Return Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
sku_data['ROAS'] = sku_data['Sales'] / sku_data['Sponsored Products charge total']
marketing_efficiency_summary = sku_data.groupby('ASIN').agg({
    'Sales': 'sum',
    'Sponsored Products charge total': 'sum',
    'ROAS': 'mean'
}).reset_index().sort_values(by='ROAS', ascending=False)
top_roas_skus = marketing_efficiency_summary.head(5)
bottom_roas_skus = marketing_efficiency_summary.tail(5)
```

```python
top_roas_skus
```

```python
bottom_roas_skus
```

```python
# ROAS Visualization
plt.figure(figsize=(10, 6))
plt.bar(top_roas_skus['ASIN'], top_roas_skus['ROAS'], color='red')
plt.xlabel('ASIN')
plt.ylabel('ROAS')
plt.title('Top 5 SKUs by ROAS')
plt.xticks(rotation=45)
plt.show()
```

```python
sku_data['Total COGS'] = sku_data['Cost of goods sold per unit'] * sku_data['Units sold']
sku_data['Profit'] = sku_data['Net proceeds total'] - sku_data['Total COGS']
sku_data['Profit Margin'] = (sku_data['Profit'] / sku_data['Net proceeds total']) * 100
```

```python
profit_summary = sku_data.groupby('ASIN').agg({
    'Net proceeds total': 'sum',
    'Total COGS': 'sum',
    'Profit': 'sum',
    'Profit Margin': 'mean'
}).reset_index().sort_values(by='Profit', ascending=False)
```

```python
top_profit_skus = profit_summary.head(5)
bottom_profit_skus = profit_summary.tail(5)
```

```python
top_profit_skus
```

```python
bottom_profit_skus
```

```python
# Profit Margin Visualization
plt.figure(figsize=(10, 6))
plt.bar(top_profit_skus['ASIN'], top_profit_skus['Profit Margin'], color='red')
plt.xlabel('ASIN')
plt.ylabel('Profit Margin (%)')
plt.title('Top 5 SKUs by Profit Margin')
plt.show()
```

```python
profitability_analysis = sku_data.groupby('ASIN').agg({
    'Net proceeds total': 'sum',
    'Total COGS': 'sum',
    'Profit': 'sum',
    'Profit Margin': 'mean'
}).reset_index().sort_values(by='Profit', ascending=False)
```

```python
high_cogs_skus = profitability_analysis[profitability_analysis['Total COGS'] > profitability_analysis['Net proceeds total']]
```
